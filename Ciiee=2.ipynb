```python
import statistics
age_mean=df['age'].mean()
age_median=df['age'].median()
age_mode=statistics.mode(df['age'])
fare_mean=df['fare'].mean()
fare_median=df['fare'].median()
fare_mode=statistics.mode(df['fare'])
print("Mean of age is :",age_mean)
print("Median of age is :",age_median)
print("Mode value of age is :",age_mode)
print("Mean value of Fare is :",fare_mean)
print("Median of Fare is :",fare_median)
print("Mode value of Fare is :",fare_mode)
```

```python
import matplotlib.pyplot as plt
df['survived'].replace({0:'not survived',1:' survived'},inplace= True)
survived_pclass_count = df.groupby(['survived', 'pclass']).size().unstack()
print(survived_pclass_count)
survived_pclass_count.plot(kind='bar')
plt.title('Survived vs. Pclass')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')
dd = df.isnull().sum()
print("Missing values before filling:\n",dd)
numeric_means = df.select_dtypes(include=['number']).mean()
dt = df.fillna(numeric_means)
dd1 = dt.isnull().sum()
print("Missing values after filling:\n",dd1)
```

```python
from scipy import stats
import seaborn as sns
sns.boxplot(x=df['fare'])
plt.title('Box Plot of Age')
plt.show()
z_score=stats.zscore(df['fare'])
threshold=3
outliers=df[abs(z_score)>threshold]
print("outliers are:")
print(outliers)
```
